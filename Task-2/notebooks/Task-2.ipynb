```python
dataset['SMA'] = dataset.groupby('stock')['Close'].transform(lambda x: talib.SMA(x, timeperiod=20))
dataset['EMA'] = dataset.groupby('stock')['Close'].transform(lambda x: talib.EMA(x, timeperiod=20))
dataset['MACD'] = dataset.groupby('stock')['Close'].transform(lambda x: talib.MACD(x, fastperiod=12, slowperiod=26, signalperiod=9)[0])
dataset['MACD_Signal'] = dataset.groupby('stock')['Close'].transform(lambda x: talib.MACD(x, fastperiod=12, slowperiod=26, signalperiod=9)[1])
dataset['MACD_Hist'] = dataset.groupby('stock')['Close'].transform(lambda x: talib.MACD(x, fastperiod=12, slowperiod=26, signalperiod=9)[2])
dataset['RSI'] = dataset.groupby('stock')['Close'].transform(lambda x: talib.RSI(x, timeperiod=14))
dataset['WMA'] = dataset.groupby('stock')['Close'].transform(lambda x: talib.WMA(x, timeperiod=20))
dataset['WCLPRICE'] = dataset.groupby('stock').apply(lambda x: talib.WCLPRICE(x['High'], x['Low'], x['Close'])).reset_index(level=0, drop=True)
dataset['Daily_Return'] = dataset.groupby('stock')['Close'].transform(lambda x: x.pct_change())
dataset['Volatility'] = dataset.groupby('stock')['Daily_Return'].transform(lambda x: x.rolling(window=30).std() * np.sqrt(252))
```

```python
dataset.head(100)
```

```python
dataset = dataset.dropna(subset=['SMA', 'EMA', 'MACD', 'MACD_Signal', 'MACD_Hist', 'RSI', 'WMA', 'WCLPRICE',
                                 'Daily_Return', 'Volatility'])
```

```python
sns.set(style="darkgrid")
def plot_technical_indicator(stock_name, dataset):
    stock_data = dataset[dataset['stock'] == stock_name]

    plt.figure(figsize=(14, 8))

    plt.subplot(3, 2, 1)
    plt.plot(stock_data['Date'], stock_data['Close'], label='Close Price')
    plt.plot(stock_data['Date'], stock_data['SMA'], label='SMA (20)')
    plt.plot(stock_data['Date'], stock_data['EMA'], label='EMA (20)')
    plt.title(f'{stock_name} - SMA & EMA')
    plt.legend()

    plt.subplot(3, 2, 2)
    plt.plot(stock_data['Date'], stock_data['RSI'], label='RSI (14)')
    plt.axhline(y=70, color='r', linestyle='--')
    plt.axhline(y=30, color='g', linestyle='--')
    plt.title(f'{stock_name} - RSI')
    plt.legend()

    plt.subplot(3, 2, 3)
    plt.plot(stock_data['Date'], stock_data['MACD'], label='MACD (12,26)')
    plt.plot(stock_data['Date'], stock_data['MACD_Signal'], label='MACD Signal (9)')
    plt.bar(stock_data['Date'], stock_data['MACD_Hist'], label='MACD Hist', color='gray', alpha=0.3)
    plt.title(f'{stock_name} - MACD')
    plt.legend()

    plt.subplot(3, 2, 4)
    plt.plot(stock_data['Date'], stock_data['WCLPRICE'], label='WCLPRICE')
    plt.title(f'{stock_name} - WCLPRICE')
    plt.legend()

    plt.subplot(3, 2, 5)
    plt.plot(stock_data['Date'], stock_data['Volatility'], label='Volatility (30)')
    plt.title(f'{stock_name} - Volatility')
    plt.legend()

    plt.subplot(3, 2, 6)
    plt.plot(stock_data['Date'], stock_data['Daily_Return'], label='Daily Return')
    plt.title(f'{stock_name} - Daily Return')
    plt.legend()

    plt.tight_layout()
    plt.show()
```
